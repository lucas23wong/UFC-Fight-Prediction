# 01 - Data Cleaning and Feature Engineering

This notebook cleans the UFC fight dataset in `raw/` and engineers differential features (reach, height, record) to fully prepare data for model training. The processed data is saved in `data/processed/`.

```python
# Setup and imports 
import pandas as pd 
import numpy as np 

ufc_data = pd.read_csv('../data/raw/ufc-master.csv')
ufc_data.info()
ufc_data.head()
```

We want to eliminate duplicated entries and drop all the separate weight division ranking columns. We will keep the 'better_rank' column only for our cleaned data set. 

```python
# Check for and eliminate duplicate entries as well as entries containing empty values
ufc_data.duplicated().any() 

# Drop individual weight rankings and consolidate with better_rank column
rank_cols = [col for col in ufc_data.columns if 'rank' in col.lower() and col != 'better_rank']
ufc_data_cleaned = ufc_data.drop(columns=rank_cols)
[c for c in ufc_data_cleaned.columns if 'rank' in c.lower()]

ufc_data_cleaned.head()
```

We can now also drop unecessary data for the models such as location and country. The dataframe already contains reach, age, height, KO, submission differentials. 

```python
# Remove location and country columns
remove_cols = ['location', 'country']
ufc_data_cleaned['date'] = pd.to_datetime(ufc_data_cleaned['date'])
ufc_data_cleaned.drop(columns=remove_cols, inplace=True)

# Add binary Red_Winner column 
ufc_data_cleaned['Red_Win'] = (ufc_data_cleaned['Winner'] == 'Red').astype(int)
```

The dataset is now ready for data analysis and model training. We have condensed the ranking columns into better_rank and eliminated unecessary fight information. There are 7177 fights recorded in this dataset from 2010-03-21 to 2026-03-28. Now, we can load the cleaned dataset into `data/processed/`.

```python
# Load cleaned data into data/processed/
ufc_data_cleaned.to_csv('../data/processed/cleaned_data.csv', index=False)
```
